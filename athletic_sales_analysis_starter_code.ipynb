```python
# Import Libraries and Dependencies
import pandas as pd
import numpy as np
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
Sa20_path=('Resources/athletic_sales_2020.csv')
Sa21_path=('Resources/athletic_sales_2021.csv')
Sa20_pd=pd.read_csv(Sa20_path,index_col='retailer_id')
Sa21_pd=pd.read_csv(Sa21_path,index_col='retailer_id')
```

```python
# Display the 2021 sales DataFrame
Sa21_pd.head()
```

```python
# Check the 2021 sales data types.
Sa21_pd.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
JoinPd=pd.concat([Sa20_pd,Sa21_pd], axis="rows",
    join="inner")
JoinPd.reset_index(drop=True)
```

```python
# Check if any values are null.
JoinPd.isnull().sum()
```

```python
# Check the data type of each column
JoinPd.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
JoinPd['invoice_date']=pd.to_datetime(JoinPd['invoice_date'], yearfirst=True)

#converted_JoinPd = JoinPd.copy()
#converted_JoinPd['invoice_date'] = converted_JoinPd.loc[:, 'invoice_date'].astype(float)
```

```python
# Confirm that the "invoice_date" data type has been changed.
JoinPd.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

SortPd=JoinPd.groupby(['region','state','city'])[['units_sold']].sum()
#SortPd_toflate=SortPd.copy
SortPd.columns=SortPd.columns.to_flat_index()
SortPd=SortPd.rename(columns={'units_sold':'Total_Products_Sold'})


# Show the top 5 results.
SortPd=SortPd.sort_values(by=['Total_Products_Sold'],ascending=False)

SortPd.reset_index(drop=True)
SortPd.head(5)
#topsale=SortPd.iloc[0,:]    
            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
PivPd=JoinPd.copy()
SortPp=pd.pivot_table(PivPd, index=['region','state','city'],values='units_sold', aggfunc='sum', sort=False,)

# Rename the "units_sold" column to "Total_Products_Sold"
SortPp.columns=SortPp.columns.to_flat_index()
SortPp=SortPp.rename(columns={'units_sold':'Total_Products_Sold'})
# Show the top 5 results.

SortPp.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
SalePd=JoinPd.groupby(['region','state','city'])[['total_sales']].sum()
SalePd.columns=SalePd.columns.to_flat_index()
SalePd=SalePd.rename(columns={'total_sales':'Total Sales'})
# Show the top 5 results.
SalePd=SalePd.sort_values(by=['Total Sales'],ascending=False)

SalePd.reset_index(drop=True)
SalePd.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

SalesPivPd=JoinPd.copy()
SalesSortPp=pd.pivot_table(SalesPivPd, index=['region','state','city'],values='total_sales', aggfunc='sum', sort=False).sort_values(by=['total_sales'],ascending=False)
# Optional: Rename the "total_sales" column to "Total Sales"

SalesSortPp.columns=SalesSortPp.columns.to_flat_index()
SalesSortPp=SalesSortPp.rename(columns={'total_sales':'Total Sales'})
# Show the top 5 results.
SalesSortPp.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
TimeFramePd=JoinPd.copy().loc[JoinPd['invoice_date']<'2020']
WomenSales=JoinPd.copy().loc[JoinPd['product'].str.contains('Women')&JoinPd['product'].str.contains('Footwear')&JoinPd['product'].str.contains('Athletic')]
#WomenSales=WomenAllSales.loc[WomenAllSales['product'].str.contains('Footwear')]
WomenSales
TimeFramePd #it seems my data has some problem 
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
WSortPd=WomenSales.groupby(['retailer','region','state','city'])[['units_sold']].sum()
#SortPd_toflate=SortPd.copy
WSortPd.columns=WSortPd.columns.to_flat_index()
WSortPd=WSortPd.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})


# Show the top 5 results.
WSortPd=WSortPd.sort_values(by=['Womens_Footwear_Units_Sold'],ascending=False)

WSortPd.reset_index(drop=True)
WSortPd.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
WPSalesPivPd=WomenSales.copy()
WPSalesPivPd=pd.pivot_table(WPSalesPivPd, index=['retailer','region','state','city'],values='units_sold', aggfunc='sum',sort=False).sort_values(by=['units_sold'],ascending=False)

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"

WPSalesPivPd.columns=WPSalesPivPd.columns.to_flat_index()
WPSalesPivPd=WPSalesPivPd.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})
# Show the top 5 results.
WPSalesPivPd.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
BestSaleDay=WomenSales.copy()
BestSaleDay=pd.pivot_table(BestSaleDay, index=['invoice_date'],values='total_sales', aggfunc='sum', sort=False).sort_values(by=['total_sales'],ascending=False).head(10)

# Optional: Rename the "total_sales" column to "Total Sales"

BestSaleDay.columns=BestSaleDay.columns.to_flat_index()
BestSaleDay=BestSaleDay.rename(columns={'total_sales':'Total Sales'})
WomenSales['invoice_date']=pd.to_datetime(WomenSales['invoice_date'], yearfirst=True)
# Show the table.
BestSaleDay
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
#dates=pd.date_range(WomenSales['invoice_date'].min(),WomenSales['invoice_date'].max(),freq='W')


# Sort the resampled pivot table in ascending order on "Total Sales".
ResampleBestSaleDay=pd.pivot_table(BestSaleDay, index=['invoice_date'],values='Total Sales', aggfunc='sum', sort=False).sort_values(by=['Total Sales'],ascending=False)
ResampleBestSaleDay
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
ResampleWeekly=ResampleBestSaleDay.copy()
#dates=pd.date_range(JoinPd['invoice_date'].min(),JoinPd['invoice_date'].max(),freq='W')
#JoinPdTest=pd.DataFrame(ResampleWeekly,index=pd.MultiIndex.from_product(dates))

#ResampleWeekly.columns = pd.date_range(ResampleWeekly['invoice_date'].min(),periods=24,freq='W')
ResampleWeekly=ResampleBestSaleDay.resample('W').sum()#.sort_values(by=['Total Sales']).head(10)

# Sort the resampled pivot table in ascending order on "Total Sales".
ResampleWeekly.sort_values(by=['Total Sales'],ascending=False).head(10)
```
